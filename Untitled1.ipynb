```python
basefolder= r"H:\Data\raw\7633_recall"
base_path = Path(basefolder)



import json
from pathlib import Path
import numpy as np
import pandas as pd
bc_tsv=r"H:\Data\raw\7633\7633_recall\sorted\sorter_output\bombcell\cluster_bc_unitType.tsv"
root = Path(r"H:\Data\raw")
npy = root / "7633" / "7633_recall" / "analyzer_clean" / "sorting" /  "properties" / "original_cluster_id.npy"

out = root / "7633" / "7633_recall" / "analyzer_clean"  / "spikeinterface_gui" / "curation_data.json"
out.parent.mkdir(parents=True, exist_ok=True)

si = np.load(npy)                # 1D array: index = phy cluster_id, value = si_unit_id
map_df = pd.DataFrame({"cluster_id": np.arange(len(si)), "si_unit_id": si})
bc_df = pd.read_csv(bc_tsv, sep="\t")

merged = map_df.merge(bc_df, on="cluster_id", how="left")

manual = []
opts = set()
for _, r in merged.iterrows():
    si_id = r.get("si_unit_id")
    lbl = r.get("bc_unitType")
    if pd.isna(si_id) or pd.isna(lbl):
        continue
    try:
        uid = int(si_id)
    except Exception:
        continue
    lbl = str(lbl)
    opts.add(lbl)
    manual.append({"unit_id": uid, "bombcell": [lbl]})  # v1 style

cur = {
    "format_version": "2",
    "unit_ids": sorted({int(x) for x in si if not np.isnan(x)}),
    "merges": [],
    "splits": [],
    "removed": [],
    "manual_labels": manual,
    "label_definitions": {
        "bombcell": {"exclusive": True, "label_options": sorted(opts)}
    },
}

with open(out, "w", encoding="utf8") as f:
    json.dump(cur, f, indent=2)

print("Wrote v1-style curation to", out)
```
