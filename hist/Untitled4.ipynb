```python
import os
from pathlib import Path

import shutil


def concat_csv_fast(input_dir, output_file):
    input_dir = Path(input_dir)
    output_file = Path(output_file)

    csv_files = sorted(input_dir.glob("*.csv"))
    if not csv_files:
        print("No CSV files found.")
        return

    with output_file.open("wb") as out_f:
        # Write header from first file
        with csv_files[0].open("rb") as f:
            header = f.readline()
            out_f.write(header)

            # Copy body of first file
            shutil.copyfileobj(f, out_f, length=1024 * 1024)

        # Copy all subsequent files without their headers
        for csv_path in csv_files[1:]:
            with csv_path.open("rb") as f:
                f.readline()  # skip header
                shutil.copyfileobj(f, out_f, length=1024 * 1024)

    print(f"Done. Combined: {len(csv_files)} files → {output_file}")


if __name__ == "__main__":
    concat_csv_fast(
        input_dir=r"D:\Data\aligned\wrangled\chunks",
        output_file=r"D:\Data\aligned\wrangled\combined_data.csv"
    )
```
