```python
import os

# List of environment variables to check
parallel_vars = [
    "MKL_NUM_THREADS",
    "OPENBLAS_NUM_THREADS",
    "OMP_NUM_THREADS",
    "BLOSC_NTHREADS",
    "BLOSC_MIN_DATA_SIZE",
    "NUMBA_NUM_THREADS",
]

for var_name in parallel_vars:
    # Use os.environ.get() to safely retrieve the variable's value
    # It returns None if the variable is not set, preventing errors.
    value = os.environ.get(var_name)

    if value is not None:
        print(f"**{var_name}**: Set to **{value}**")
    else:
        print(f"**{var_name}**: Not explicitly set in the environment.")
```
