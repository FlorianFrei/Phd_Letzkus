```python
import spikeinterface.full as si
import pandas as pd
from pathlib import Path
import re

def export_unit_metadata(base_directory):
    base_path = Path(base_directory)
    
    if not base_path.exists():
        print(f"Error: {base_directory} not found.")
        return

    for folder in base_path.iterdir():
        if folder.is_dir() and (folder / "spike_times").is_dir():
            
            meta_dir = folder / "Meta"
            meta_dir.mkdir(exist_ok=True)
            output_path = meta_dir / "unit_extremum_metadata.csv"
            
            if output_path.exists():
                print(f"Skipping {folder.name}: CSV already exists.")
                continue
            
            print(f"Processing: {folder.name}...")
            try:
                folder = Path(r"H:\Data\raw\8155_naive_g0")
                an_folder = folder / "analyzer"
                analyzer = si.load_sorting_analyzer(an_folder)
                print(1)
                
                
                # 1. Get extremum channels
                extremum_channels = si.get_template_extremum_channel(analyzer)
                print(2)
                # 2. Get channel locations directly from the analyzer
                # Returns a numpy array of [x, y]
                locations = analyzer.get_channel_locations()
                chan_ids = analyzer.channel_ids
                print(3)
                
                # Map channel_id to the Y-coordinate (index 1)
                location_dict = {chan_ids[i]: locations[i, 1] for i in range(len(chan_ids))}
                
                data = []
                for unit_id, ch_id in extremum_channels.items():
                    # Extract numeric ID (imec0.ap#AP376 -> 376)
                    match = re.findall(r'\d+', str(ch_id))
                    clean_ch = match[-1] if match else ch_id
                    
                    # Get distance from tip
                    dist_from_tip = location_dict.get(ch_id, 0)
                    
                    data.append({
                        'unit_id': unit_id,
                        'extremum_channel': clean_ch,
                        'dist_from_tip_um': dist_from_tip
                    })
                
                df = pd.DataFrame(data)
                df.to_csv(output_path, index=False)
                print(f"Successfully saved to Meta folder.")
                
            except Exception as e:
                print(f"Failed to process {folder.name}: {e}")

if __name__ == "__main__":
    DIR_TO_SCAN = r"H:\Data\raw"
    export_unit_metadata(DIR_TO_SCAN)
```

```python
import shutil
from pathlib import Path

def consolidate_metadata_files(root_directory, destination_directory):
    root_path = Path(root_directory)
    dest_path = Path(destination_directory)
    
    # Create the target directory if it doesn't exist
    dest_path.mkdir(parents=True, exist_ok=True)
    
    # Iterate through the Meta folders
    # This glob looks for the specific file we created in the previous step
    for meta_file in root_path.glob("*/Meta/unit_extremum_metadata.csv"):
        # The parent folder of 'Meta' is the recording folder
        folder_name = meta_file.parent.parent.name
        
        # Enforce the x_y naming convention
        # split('_') breaks "8155_recall_g0" into ["8155", "recall", "g0"]
        parts = folder_name.split('_')
        if len(parts) >= 2:
            new_name_base = f"{parts[0]}_{parts[1]}"
        else:
            new_name_base = folder_name # Fallback if no underscores exist
            
        new_filename = f"{new_name_base}.csv"
        target_file_path = dest_path / new_filename
        
        try:
            shutil.copy2(meta_file, target_file_path)
            print(f"Copied: {folder_name} -> {new_filename}")
        except Exception as e:
            print(f"Error copying {folder_name}: {e}")

if __name__ == "__main__":
    SOURCE_DIR = r"H:\Data\raw"
    OUTPUT_DIR = r"H:\Data\aligned\new\extra"
    consolidate_metadata_files(SOURCE_DIR, OUTPUT_DIR)
```
